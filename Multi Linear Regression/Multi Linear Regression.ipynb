```python


import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
# Import Dataset from sklearn
from sklearn.datasets import load_boston
import math
import matplotlib.pyplot as plt


# Load Boston Dataset
boston = load_boston()
# Creating pd DataFrames from boston dataset
boston_df = pd.DataFrame(data= boston.data, columns= boston.feature_names)
target_df = pd.DataFrame(data= boston.target, columns= ['prices']) # Concatenate the DataFrames
boston_df = pd.concat([boston_df, target_df], axis= 1)
#X and Y variables
x= boston_df.drop(labels= 'prices', axis= 1)
y= boston_df['prices']
#print(x)
#print(y)

# Splitting the Dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.35, random_state= 60)
#print(X_train)
#print(y_train)
# Instantiating LinearRegression() Model
lr = LinearRegression()
# Training/Fitting the Model
lr.fit(x_train, y_train)

# Making Predictions
# lr.predict(x_test)
print()
print("1.List the x value and the predicted y value")
print()
print("The x values are:\n")
print(x_test)
pred = lr.predict(x_test)
print()
print("\nPredicted y values:\n")
print(pred)
print()
#Regression line
print()

print('The intercept for the regression line is: ',lr.intercept_)
print()
print('The slope for the regression line is: ',lr.coef_)
print()
lineeq=lr.coef_*x+lr.intercept_
print("2. Scatter plot with regression line")
plt.plot(x,lineeq, color = "m")
plt.scatter(x.RAD, y,color = "r",marker = "o", s = 10)
plt.scatter(x.AGE, y,color = "g",marker = "o", s = 10)
plt.plot(x,lineeq, color = "r")
plt.xlabel('X')
plt.ylabel('Y')

# Evaluating Model's Performance
print()
print('3. Display the error')
print('Mean Absolute Error:', mean_absolute_error(y_test, pred))
print('Mean Squared Error:', mean_squared_error(y_test, pred))
print('Mean Root Squared Error:', np.sqrt(mean_squared_error(y_test, pred)))
```
